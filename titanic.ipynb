13 to 19= teenager\n
19-30=Young\n
30-59=Adult\n
 >=60 = Senior citizen

```python
age_group=[]
for i in df1["Age"]:
    if(i<13):
        age_group.append("Child")
    elif(i>=13 and i<=19):
        age_group.append("Teenager")
    elif(i>19 and i<=30):
        age_group.append("Young")
    elif(i>30 and i<60):
        age_group.append("Adult")
    else:
        age_group.append("Senior Citizen")
```
